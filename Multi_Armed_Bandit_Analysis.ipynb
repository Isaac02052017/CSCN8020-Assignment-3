```python
# Import libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set(style="whitegrid")
```

```python
# Load CSV results
stationary = pd.read_csv("submissions_round1.csv")
non_stationary = pd.read_csv("submissions_round2.csv")

print("Stationary shape:", stationary.shape)
print("Non-Stationary shape:", non_stationary.shape)
stationary.head()
```

```python
print(stationary.columns)
```

```python
# Plot average reward for Stationary Casino
plt.figure(figsize=(10,5))
stationary['average_reward'] = stationary['total_reward'] / stationary['steps']
plt.plot(stationary['steps'], stationary['average_reward'], label='Stationary Casino', color='blue')
plt.title("Average Reward â€“ Stationary Environment")
plt.xlabel("Steps"); plt.ylabel("Average Reward"); plt.legend(); plt.show()
```

```python
# Plot average reward for Non-Stationary Casino
plt.figure(figsize=(10,5))
non_stationary['average_reward'] = non_stationary['total_reward'] / non_stationary['steps']
plt.plot(non_stationary['steps'], non_stationary['average_reward'], label='Non-Stationary Casino', color='orange')
plt.title("Average Reward â€“ Non-Stationary Environment")
plt.xlabel("Steps"); plt.ylabel("Average Reward"); plt.legend(); plt.show()
```

```python
# Compare both environments
plt.figure(figsize=(10,6))
plt.plot(stationary['steps'], stationary['average_reward'], label='Stationary', color='blue')
plt.plot(non_stationary['steps'], non_stationary['average_reward'], label='Non-Stationary', color='red')
plt.title("Exploration vs Exploitation Comparison")
plt.xlabel("Steps"); plt.ylabel("Average Reward"); plt.legend(); plt.show()
```

## Reflection 1 â€“ Exploration vs Exploitation

- Exploitation â†’ choosing best-known arm for reward.
